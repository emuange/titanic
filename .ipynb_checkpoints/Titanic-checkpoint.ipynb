# Preprocessing

```python
train.columns.values
```

```python
train.head()
```

```python
train.info()
```

```python
train['Name'].unique().tolist()
```

```python
titles = ['Mrs', 'Mr', 'Don', 'Jonkheer', 'Master', 'Miss', 'Major', 
         'Rev', 'Dr', 'Ms', 'Mlle', 'Col', 'Capt', 'Mme', 'Countess']

def get_title(name):
    for title in titles:
        if title in name:
            return title
    return 'None'
```

```python
train['Title'] = train['Name'].apply(lambda x: get_title(x))
train.head()
```

```python
train['Age'].mean()
```

```python
# Fill empty rows with mean age
train['Age'].fillna(train['Age'].mean(), inplace = True)
train.head()
```

```python
train.drop('Cabin', axis = 1, inplace = True)
train.head()
```

```python
train.info()
```

```python
train.dropna(inplace = True)
train.info()
```

```python
x_train = train[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked', 'Title']]
y_train = train['Survived']
```

```python
x_train.head()
```

```python
x_train = pd.get_dummies(x_train)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
x_train.info()
```

# Model Training

```python
# Initializing the model
clf = RandomForestClassifier(max_depth = 75, random_state = 0)
```

```python
# Training the Model
clf.fit(x_train, y_train)
```

# Model Testing
<h2>Preprocessing</h2>

```python
test['Title'] = test['Name'].apply(lambda x: get_title(x))
test['Age'].fillna(test['Age'].mean(), inplace = True)
test.drop('Cabin', axis = 1, inplace = True)
test.head()
```

```python
test['Fare'].fillna(test['Fare'].mean(), inplace = True)
test.info()
```

```python
x_test = test[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked', 'Title']]
x_test.head()
```

```python
x_test = pd.get_dummies(x_test)
```

```python
x_test.info()
```

```python
total_titles = train['Title'].unique().tolist()
test_titles = test['Title'].unique().tolist()

for title in total_titles:
    if title not in test_titles:
        print('Title ' + str(title) + ' not in Test Set')
        x_test['Title_' + str(title)] = 0
```

```python
x_test.info()
```

```python
x_test
```

```python
list(x_test.columns)
```

```python
list(x_train.columns)
```

```python
x_test = x_test[['Pclass',
 'Age',
 'SibSp',
 'Parch',
 'Fare',
 'Sex_female',
 'Sex_male',
 'Embarked_C',
 'Embarked_Q',
 'Embarked_S',
 'Title_Capt',
 'Title_Col',
 'Title_Countess',
 'Title_Don',
 'Title_Dr',
 'Title_Jonkheer',
 'Title_Major',
 'Title_Master',
 'Title_Miss',
 'Title_Mlle',
 'Title_Mme',
 'Title_Mr',
 'Title_Mrs',
 'Title_Ms',
 'Title_Rev']]
```

```python
list(x_test.columns)
```

```python
Survived = clf.predict(x_test)
Survived
```

# Generating the Submission File

```python
output = pd.DataFrame({'PassengerId': test.PassengerId, 'Survived': Survived})
output.head()
```
